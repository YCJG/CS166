```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
# Task 1a

results = []
sample_path=[]
for i in range(1000): #1000 walks
    position=[0] #starting at point 0
    for j in range(100): #100 steps each walk
        if position[-1] == -4: #we have a hard barrier at -4
            position.append(-3) #so next step is -3
        else:
            position.append(position[-1] + np.random.choice([-1, 1]))
    results.append(position[-1]) #record final position
    sample_path.append(position)
```

```python
plt.hist(results)
```

```python
plt.plot(sample_path[0])
```

```python
# Task 1b

results2 = []

for i in range(1000): #1000 walks
    position=[0] #starting at point 0
    for j in range(100): #100 steps each walk
        if position[-1] == 6:
            p = random.random()
            if p <0.25:
                position.append(7)
            else:
                position.append(5)
        else:
            position.append(position[-1] + np.random.choice([-1, 1]))
    results2.append(position[-1]) #record final position
```

```python
# Task 2
results=[]
sample_path = []
for num_step in [20, 400, 1600]:
    for var in [0.5, 1, 2]:
        for i in range(1000):
            path = [[0,0]]
            for j in range(num_step):
                direction = random.random()*math.pi*2
                step = np.random.normal(0, var)
                x = step*np.cos(direction)
                y = step*np.sin(direction)
                position = []
                position.append(path[-1][0]+x)
                position.append(path[-1][1]+y)
                path.append(position)
            if num_step==20 and var == 0.5:
                sample_path.append(path)
            results.append(path[-1])
```

```python
x=[]
y=[]
for i in sample_path[0]:
    x.append(i[0])
    y.append(i[1])
plt.figure()
plt.plot(x, y, 'k-')
plt.axis('square')
```

```python
# density
sample = results[:1000]

x2 = []
y2= []
for i in sample:
    x2.append(i[0])
    y2.append(i[1])
plt.figure()
plt.plot(x2,y2, 'k.', markersize=1)
plt.axis('square')
```

```python
# Task 3
results=[]
sample_path = []
bound_count=0
hole_count=0
for i in range(1000):
    path = [[0,0]]
    for j in range(100): #100 steps
        x = path[-1][0] + np.random.choice([-1,1]) #arbitrary step size of 1 since not given in prompt
        y = path[-1][1] + np.random.choice([-1,1])
        path.append([x,y])
        if x >= 20 or x <= -20 or y >= 20 or y <= -20:
            bound_count+=1
            break
        elif 13<=x<=14 and 7<=y<=8:
            hole_count +=1
            break
        else:
            pass
    results.append(path[-1])
```

```python
print(hole_count)
print(hole_count/1000) #fraction of entering the hole
```

```python
# last positions
x3 = []
y3= []
for i in results:
    x3.append(i[0])
    y3.append(i[1])
plt.figure()
plt.plot(x3,y3, 'k.', markersize=1)
plt.axis('square')
```
