# Solution to pre-class work

If $q$ is the probability of a node not being in the LCC and if there are $n$ nodes in the network, then the number of nodes in the LCC will be $n(1-q)$.

```python
def calculate_q(k):
    '''
    Use a numerical root finder to determine q from the equation
    q = exp(k*(q-1)).
    '''
    from scipy.optimize import root
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of nodes
nodes = 1000

# Average degree
k = np.linspace(1, 10, 20)

# Probability of not being in the LCC
q = np.array([calculate_q(_) for _ in k])

# Theoretical size of the LCC
LCC_size = nodes * (1 - q)

plt.figure(figsize=(8, 6))
plt.plot(k, LCC_size, 'k.')
plt.xlabel('average degree')
plt.ylabel('size of LCC')
plt.title('Theoretical LCC size vs average degree')
plt.show()
```
