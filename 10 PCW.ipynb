```python
# Exercise 8

import numpy as np
import matplotlib.pyplot as plt

p_head = 0.6

coin_flips = np.array(np.random.uniform(0, 1, size=30) < p_head, dtype=int) * 2 - 1

plt.figure(figsize=(10, 5))
plt.plot(np.cumsum(coin_flips))

plt.title('Path')
plt.xlabel('Number of coin flips')
plt.ylabel('Cumulative value of outcomes')

plt.show()
```

```python
results = []

for i in range(200):
    coin_flips = np.array(np.random.uniform(0, 1, size=30) < p_head, dtype=int) * 2 - 1
    final_position = sum(coin_flips)
    results.append(final_position)

plt.figure(figsize=(10, 5))
plt.hist(results)
plt.axvline(x=np.mean(results), color = 'black', label = 'Experimental mean')
plt.axvline(x=0.2 * 30, color='red', label = 'Theoretical mean')

plt.title('Simulation Results')
plt.xlabel('Cumulative value of outcomes')
plt.ylabel('Count')
plt.legend()

plt.show()
```

```python
# Exercise 10

def gambler_ruin(upper_bound):
    gambler = 100
    house = 2000

    fortune = []
    num_gambles = upper_bound
    
    for i in range(upper_bound):
        trial = np.random.choice([1, -1], size = 1, p = [0.5, 0.5])
        if trial ==1:
            gambler += 1
            house -+ 1
        else:
            gambler -= 1
            house += 1

        fortune.append(gambler)

        if gambler == 0 or house == 0:
            num_gambles = i
            break
            
    plt.plot(range(num_gambles), fortune)
    plt.xlim(0, 900)
    plt.xlabel('Number of gambles')
    plt.ylabel('Cumulative fortune')
    plt.show()
    

for gamble in [100, 300, 500, 700, 900]:
    gambler_ruin(gamble)
```
